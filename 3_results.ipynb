```python
import pandas as pd
import numpy as np
import warnings
import json
import os

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
# loop through every score json in scores directory
# and load them into dataframe

directory = os.fsencode('scores')
res = pd.DataFrame()
    
for file in os.listdir(directory):
    with open(f'scores/{file.decode()}') as f:
        j = json.load(f)
        res = pd.concat([res, pd.DataFrame(j)], axis=1)

data = res.T.reset_index()
data.sort_values(by=['f1'], ascending=False)
```
